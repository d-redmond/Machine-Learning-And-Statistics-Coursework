### Note that there are the same number of ways to get 4 tails as there to get 4 heads. Explain why this is.

### Plot bar charts or histograms of any three different distributions listed at the following link:

https://numpy.org/doc/stable/reference/random/generator.html#distributions

The outcome of whether flipping a coin lands on heads or tails is mutually exclusive. Heads is no more likely than tails. 

Each flip has two possible outcomes. Flipping a coin four times has sixteen possible outcomes. Each of those sixteen possible outcomes will be equally likely to be heads or tails. 

The probability of getting four heads or four tails with a consecutive four flips is 1/16 (2x2x2x2=16).

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Uniform/Constant distribution
rng = np.random.default_rng(0)

# Generating 500 numbers on a uniform distribution.
u_samples = rng.uniform(-1, 2, 500)
bin = np.arange(-2,3,0.1)
plt.hist(u_samples, bins=bin, edgecolor='black') 
plt.title('Uniform/Constant') 

plt.show()
```

```python
# Frequency Distribution
%matplotlib inline
plt.rcParams.update({'figure.figsize':(8,4), 'figure.dpi':100})
x = np.random.normal(size = 1300)
plt.hist(x, bins=50)
plt.gca().set(title='Frequency Distribution', ylabel='Y Label');
```

```python
# Bar stacked with pandas
import pandas as pd

plotdata = pd.DataFrame({
    "Type A":[57,67,77,83],
    "Type B":[68,73,80,79],
    "Type C":[73,78,80,85]},
    index=["blue", "red", "yellow", "white"])
plotdata.plot(kind='bar', stacked=True,figsize=(20, 12))

plt.title("Stacked Bar")

plt.xlabel("X Label")

plt.ylabel("Y Label")
```
